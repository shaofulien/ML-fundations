```python
import numpy as np

def E_in(weight,test):
    result = np.sign(test[:,:-1].dot(weight))
    return (result!=test[:,-1]).sum()/test.shape[0]
def gradient(y,w,x):
    temp=y*w.dot(x)
    return (-y*x)/(np.exp(temp)+1)
train=[]
test=[]
E_in_gradient=[]
E_in_SGD=[]
E_out_gradient=[]
E_out_SGD=[]

with open('hw3_train.dat') as file:
    for i in file.readlines():
        train.append([1]+list(map(float,i.strip().split(' '))))

with open('hw3_test.dat') as file:
    for i in file.readlines():
        test.append([1]+list(map(float,i.strip().split(' '))))
        
train=np.array(train)
test=np.array(test)

weight_D = train.shape[1]-1

number_train_size = train.shape[0]

weight=np.zeros(weight_D)
learning_R=0.01

for i in range(2000):
    sum_gradient=np.zeros(weight_D)
    for j in range(number_train_size):
        sum_gradient+=gradient(train[j][-1],weight,train[j][:-1])
    sum_gradient=sum_gradient/number_train_size
    weight-=learning_R*sum_gradient
    #print (weight)
    E_out_gradient.append(E_in(weight,test))
    E_in_gradient.append(E_in(weight,train))
    #print (E_in(weight,train))
    
    
```

```python
weight=np.zeros(weight_D)
learning_R=0.001

for i in range(2000):
    sum_gradient=np.zeros(weight_D)
    sum_gradient=gradient(train[i%number_train_size][-1],weight,train[i%number_train_size][:-1])
    weight-=learning_R*sum_gradient
    
    E_out_SGD.append(E_in(weight,test))
    E_in_SGD.append(E_in(weight,train))
    #print (E_in(weight,train))
```

```python
import matplotlib.pyplot as plt 
plt.plot(E_in_gradient,label='Gradient_E_in') 
plt.plot(E_in_SGD,label='SGD_E_in')
plt.legend(loc='lower left')
plt.show() 
```

```python
plt.plot(E_out_gradient,label='Gradient_E_out') 
plt.plot(E_out_SGD,label='SGD_E_out')
plt.legend(loc='lower left')
plt.show() 
```
